EDA

EDA on Gender column

```python
#check who have the most purchasing power 'woman and man'
sns.countplot(x='Gender',data=df)
```

```python
x=sns.countplot(x='Gender',data=df)
for bars in x.containers:
  x.bar_label(bars)
```

from above graph we can see that the woman has more purchasing power as compare to man

```python
sells_data=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sells_data,hue='Gender')
```

from above graph we can describe that womans are spending more money.

EDA on Gender and Orders

```python
sales_data=df.groupby(['Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Gender',y='Orders',data=sales_data,hue='Gender')
```

from above graph we can see the 'womans' ordered more products

EDA with Age Group

```python
sells_data=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in sells_data.containers:
  sells_data.bar_label(bars)
```

from above one graph we can analyze the age group from 26-35 'womans' have more purchasing power

```python
#   EDA with 'Shadi Status' column
salles_data=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in salles_data.containers:
  salles_data.bar_label(bars)
```

in above graph is clearly showing that the Married peoples 'woman' are spending more.

```python
salles_data=df.groupby(['Marital_Status'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Marital_Status',y='Orders',data=salles_data)
```

the married persons are ordered most products,clearly showing graphly

```python
#EDA with 'Marital Status' and 'Amount'
salles_data=df.groupby('Marital_Status',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=salles_data)
```

from above graph we can see the married persons are spending more money

```python
#EDA with state and counts
sns.set(rc={'figure.figsize':(28,7)})
salles_data=sns.countplot(x='State',data=df)
for bars in salles_data.containers:
  salles_data.bar_label(bars)
```

from above graph we can say that 'uttar prades','Maharashtra' and 'Karnataka' ordered more products

```python
# EDA with 'State' and 'Orders'
df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(25,10)})


sns.barplot(x='State',y='Orders',data=df)
```

```python
#DEA with occupations
sns.set(rc={'figure.figsize':(28,12)})
salles_data=sns.countplot(x='Occupation',data=df)
for bar in salles_data.containers:
  salles_data.bar_label(bar)
```

IT Sector ,HealthCare and Aviation have more perchasing power

```python
#EDA with Occupation and amount
sales_data=df.groupby('Occupation',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(x='Occupation',y='Amount',data=sales_data)
```

from above graph we can see the IT Sector ,HealthCare and Aviation Occupation are perchasing more

```python
#EDA with Product_Category
sns.set(rc={'figure.figsize':(30,17)})
sales_data=sns.countplot(x='Product_Category',data=df,hue='Gender')
for bar in sales_data.containers:
  sales_data.bar_label(bar)
```

most buying products are clothing foods and electronics

Most buyers are the woman from UttarPrades,Maharashtra and karnatka their age group belongs to 26-35 and they are from IT Sectors ,HealthCare and Aviation.
Most purchased products are Clothinf,Food and Electronics
